#### Converting yaml data to pandas dataframe

```python
def getScore(innings):
    score = {}
    for inning in range(len(innings)):
        runs = []
        numDeliveries = len(innings[inning].values()[0]['deliveries'])
        for d in range(numDeliveries):
            runs.append(innings[inning].values()[0]['deliveries'][d].values()[0]['runs']['total'])
        score[inning] = sum(runs)
    if (len(score.keys()) == 1):
        score[1] = np.NaN
    return score
```

```python
def getTeams(info):
    decision = info['toss']['decision']
    winner = info['toss']['winner']
    teams = info['teams']
    teamId = {}
    if(decision == 'field'):
        teamId[1] = winner
        teamId[0] = [t for t in teams if t != winner][0]
    else:
        teamId[0] = winner
        teamId[1] = [t for t in teams if t != winner][0]
    return teamId
```

```python
def getData(dataMap, idx):
    info = dataMap['info']
    innings = dataMap['innings']
    meta = dataMap['meta']
    
    score = getScore(innings)
    teamIds = getTeams(info)
    gameDate = info['dates'][0]
    matchType = info['match_type']
    overs = info['overs']
    venue  = info['venue']
#     player_of_match = info['player_of_match'][0]
#     tossWinner = info['toss']['winner']
#     tossDecision = info['toss']['decision']
    
    try:
        gameWinner = info['outcome']['winner']
    except:
        gameWinner = np.NaN
    
    try:
        city = info['city']
    except:
        city = np.NaN
    
    glist = [teamIds[0], teamIds[1], score[0], score[1], gameWinner, city, gameDate, matchType,
             overs, venue]#, player_of_match, tossWinner, tossDecision]
    columns = ['team0', 'team1', 'score0', 'score1', 'winner', 'location', 'date', 'match_type',
               "overs", "venue"]#, "player_of_match", "tossWinner", "tossDecision"]
    return pd.DataFrame([glist], index=[idx], columns=columns)
```

```python
allGames = glob.glob("*.yaml")

games = []
for idx, g in enumerate(allGames):
    f = open(g)
    dataMap = yaml.load(f)
    data = getData(dataMap, idx)
    games.append(data)
    f.close()
df = pd.concat(games)
```

```python
def getYear(d):
    return d.year

def getMonth(d):
    return d.month

def getDay(d):
    return d.day
```

```python
df["year"] = df["date"].apply(getYear)
df["month"] = df["date"].apply(getMonth)
df["day"] = df["date"].apply(getDay)
```

```python
allTeams = set(list(set(df.team0))+list(set(df.team1)))
```

### Average scores of teams and their opponents

```python
scoreStats = {}
for team in allTeams:
    teamScore = []
    opponentScore = []
    numGames = 0
    for i, r in df.iterrows():
        if r.team0 == team:
            if (isnan(r.score0) == False and isnan(r.score1) == False):
                teamScore.append(r.score0)
                opponentScore.append(r.score1)
                numGames += 1
        elif r.team1 == team:
            if (isnan(r.score0) == False and isnan(r.score1) == False):
                teamScore.append(r.score1)
                opponentScore.append(r.score0)
                numGames += 1
    scoreStats[team] = [numGames, int(round(sum(teamScore)/numGames)), int(round(sum(opponentScore)/numGames))]
```

```python
aScore = pd.DataFrame(scoreStats).T
aScore.columns = ['numGames', 'avgTeamScore', 'avgOpponentScore']
aScore['margin'] = aScore.avgTeamScore - aScore.avgOpponentScore
aScore.sort('margin', ascending = False)
```

```python
plt.rc('figure', figsize=(20,12))
plotScore = aScore.sort('margin', ascending = False)[['avgTeamScore', 'avgOpponentScore']]
plotScore.plot(kind="bar")
plt.ylabel('Score')
plt.title('Average Team Score vs. Average Opponent Score ')
plt.show()
```

### Team win/loss ratio

```python
stats = {}
for team in allTeams:
    numWins = int(round(sum(df.winner == team)))
    totalGames = int(round(sum(df.team0 == team) + sum(df.team1 == team)))
    winPercentage = numWins/float(totalGames)
    stats[team] = {'numWins':numWins, 'totalGames':totalGames, 'winPercentage':winPercentage}
```

```python
stats = pd.DataFrame(stats).T
stats[['numWins', 'totalGames']].astype(int)
stats = stats.sort('totalGames', ascending=False)
stats
```

```python
stats['numLosses'] = stats.totalGames * (1-stats.winPercentage)
plotstats = stats[["numWins", "numLosses"]]
```

### ranking teams!

```python
def create_team_graph(teams):
    graph = nx.DiGraph()
    for i in teams.index:
        team = teams.ix[i]
        try:
            score_diff = team['score0'] - team['score1']
        except:
            score_diff = 0

        if score_diff > 0:
            winner = team['team0']
            loser = team['team1']
            point_differential = team['score0']#float(abs(team['score0']- team['score1']))# /(team['score0']+ team['score1'])
        else:
            winner = team['team1']
            loser = team['team0']
            point_differential = team['score1']#float(abs(team['score1']- team['score0']))# /(team['score0']+ team['score1'])
        update_graph(winner, loser, point_differential, graph)
    return graph
```

```python
def update_graph(winner, loser, point_differential, graph):    
    graph.add_node(winner)
    graph.add_node(loser)
    
    if graph.has_edge(loser, winner):
        updated_weight = graph[loser][winner]['weight'] + point_differential
    else:
        updated_weight = point_differential
        
    graph.add_edge(loser, winner, weight = updated_weight)
    return graph
```

```python
# Get winner for all the years
def get_winners(data):
    years = list(unique(data.year))
    result = dict()
    graphs = dict()
    for year in years:
        games = data[data.year == year]
        team_network = create_team_graph(games)
        pagerank_teams = nx.pagerank_numpy(team_network)
        pagerank_teams = pd.DataFrame(pagerank_teams.items(), columns=['team', 'page_rank'])
        pagerank_teams = pd.DataFrame(pagerank_teams.ix[:, 'page_rank']).set_index(pagerank_teams.team)
        pagerank_teams = pagerank_teams.sort(columns='page_rank', ascending=False)
        result[year] = pagerank_teams
        graphs[year] = team_network
    return result, graphs
```

```python
winners, graphs = get_winners(df)
```

```python
def plotGraph(data):
    plt.rc('figure', figsize=(20,12))
    pos = nx.spring_layout(data, iterations=200)

    nx.draw_networkx_edges(data, pos, alpha = .05)

    nx.draw_networkx_nodes(data, pos, node_color=colors[2])

    for p in pos.values():
        p[1] += .03

    lbls = nx.draw_networkx_labels(data, pos, alpha=5, font_size=8)

    plt.xticks([])
    plt.yticks([])
```
